```python
import pygame
import random
import sys
from time import sleep
```

```python
BLACK = (0, 0, 0)
screen_width = 480
screen_height = 640
```

```python
# 게임에 등장하는 객체 드로잉
def drawObject(obj, x, y) :
    global gamepad
    gamepad.blit(obj, (x, y))

def initGame() :
    global gamepad, clock, background, fighter
    pygame.init()
    gamepad = pygame.display.set_mode((screen_width, screen_height))
    pygame.display.set_caption('PyShooting')
    background = pygame.image.load("./png/background.png").convert() # 배경화면
    background = pygame.transform.scale(background, (480, 640)) # 스크린 사이즈에 맞게 배경화면 크기 조절
    fighter = pygame.image.load('./png/pp.png') # 전투기 그림
    fighter = pygame.transform.scale(fighter, (48, 64)) # 전투기 사이즈 조절
    clock = pygame.time.Clock()
    
def runGame() :
    global gamepad, clock, background, fighter

    # 총알 생성
    gun = pygame.image.load('./png/gun.png')
    gun = pygame.transform.scale(gun, (20, 40))
    gunSize = gun.get_rect().size
    gunWidth, gunHeight = gunSize[0], gunSize[1]

    # 총알 초기 위치 설정
    gunX = random.randrange(0, screen_width - gunWidth)
    gunY = 0
    gunSpeed = 2

    # 비행기 크기
    fighterSize = fighter.get_rect().size
    fighterWidth = fighterSize[0]
    fighterHeight = fighterSize[1]

    # 전투기 초기 위치 (x,y) 설정
    x = screen_width * 0.43
    y = screen_height * 0.85
    fighterX, fighterY = 0, 0

    onGame = False
    while not onGame :
        for event in pygame.event.get() :
            if event.type in [pygame.QUIT] :
                pygame.quit()
                sys.exit()

            if event.type in [pygame.KEYDOWN] :
                if event.key == pygame.K_LEFT : # 전투기 왼쪽으로 이동
                    fighterX -= 5
                elif event.key == pygame.K_RIGHT : # 전투기 오른쪽으로 이동
                    fighterX += 5
                elif event.key == pygame.K_UP : # 전투기 위쪽으로 이동
                    fighterY -= 5
                elif event.key == pygame.K_DOWN : # 전투기 아래쪽으로 이동
                    fighterY += 5

            if event.type in [pygame.KEYUP] : # 방향키를 떼면 전투기 멈춤
                if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT :
                    fighterX = 0
                elif event.key == pygame.K_UP or event.key == pygame.K_DOWN :
                    fighterY = 0
                    
        drawObject(background, 0, 0) # 게임 화면 (검은색)

        # 전투기 위치 재조정
        x += fighterX
        if x < 0 :
            x = 0
        elif x > screen_width - fighterWidth :
            x = screen_width - fighterWidth

        y += fighterY
        if y < 0 :
            y = 0
        elif y > screen_height - fighterHeight :
            y = screen_height - fighterHeight

        drawObject(fighter, x, y)

        gunY += gunSpeed # 운석 아래로 움직임
        drawObject(gun, gunX, gunY)

        pygame.display.update() # 게임 화면을 다시 그림
        clock.tick(60) # 게임 화면의 초당 프레임 수를 60으로 설정

    pygame.quit() # pygame 종료
```

```python
initGame()
runGame()
```
