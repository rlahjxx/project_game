```python
import turtle
import math
import random

wn = turtle.Screen()
wn.bgcolor("white")
wn.title("미로게임")
wn.setup(700, 700)
wn.tracer(0)

# 저장된 이미지 파일 불러오기
turtle.register_shape("enemy2.gif")
turtle.register_shape("slime2.gif")
turtle.register_shape("cookie2.gif")
# turtle.register_shape("Cube.gif")



# create Pen
class Pen(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.shape("square")
        self.color("black")
        self.penup()
        self.speed(0)
# Player class 지정      
class Player(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.shape("slime2.gif")
        self.color("blue")
        self.penup()
        self.speed(0)
        self.gold = 0
        
    def go_up(self):
        move_to_x = player.xcor()
        move_to_y = player.ycor() + 24
        
        if (move_to_x, move_to_y) not in walls:
            self.goto(move_to_x, move_to_y)
            
        
    def go_down(self):
        move_to_x = player.xcor()
        move_to_y = player.ycor() - 24
        
        if (move_to_x, move_to_y) not in walls:
            self.goto(move_to_x, move_to_y)
    
    def go_left(self):
        move_to_x = player.xcor() - 24
        move_to_y = player.ycor()
        
        if (move_to_x, move_to_y) not in walls:
            self.goto(move_to_x, move_to_y)
    
    def go_right(self):
        move_to_x = player.xcor() + 24
        move_to_y = player.ycor()
        
        if (move_to_x, move_to_y) not in walls:
            self.goto(move_to_x, move_to_y)

        
    def is_collision(self, other):
        a = self.xcor()-other.xcor()
        b = self.ycor()-other.ycor()
        distance = math.sqrt((a ** 2) + (b ** 2) )
        
        if distance < 5:
            return True
        else:
            return False
        
    def destroy(self):
        self.goto(2000, 2000)
        self.hideturtle()

# 보상 클래스 생성
class Treasure(turtle.Turtle):
    def __init__(self, x, y):
        turtle.Turtle.__init__(self)
        self.shape("cookie2.gif")   # 이미지파일
        self.color("gold")          
        self.penup()
        self.speed(0)         
        self.gold = 100
        self.goto(x, y)
        
    def destroy(self):           # 플레이어가 먹으면 삭제
        self.goto(2000, 2000)
        self.hideturtle()

        
# 적 클래스 생성
class Enemy(turtle.Turtle):
    def __init__(self, x, y):
        turtle.Turtle.__init__(self)
        self.shape("enemy2.gif")
        self.color("red")
        self.penup()
        self.speed(0)
        self.gold = 25
        self.goto(x, y)
        self.direction = random.choice(["up", "down", "left", "right"])  # 랜덤으로 움직이는 방향 정해줌
        
    def move(self):
        if self.direction == "up":
            dx = 0
            dy = 24
        
        elif self.direction == "down":
            dx = 0
            dy = -24
            
        elif self.direction == "left":
            dx = -24
            dy = 0
            
        elif self.direction == "right":
            dx = 24
            dy = 0
        
        else:
            dx = 0
            dy = 0
        
        move_to_x = self.xcor() + dx
        move_to_y = self.ycor() + dy
        
        if (move_to_x, move_to_y) not in walls:
            self.goto(move_to_x, move_to_y)
        else:
            self.direction = random.choice(["up", "down", "left", "right"])
        
        # 움직임 타이머
        turtle.ontimer(self.move, t=random.randint(100,300))
        
        
    def destroy(self):
        self.goto(2000, 2000)
        self.hideturtle()
        
# create levels list
levels = [""]

level_1 = [
"XXXXXXXXXXXXXXXXXXXXXXXXX",
"XP XXXXXXX          XXXXX",
"X  XXXXXXX  XXXXXX  XXXXX",
"X           XXX        XX",
"XXXXXX  XX  XXXXXX  XXXXX",
"XXXXXX  XX      XX  XXXXX",
"X  XXX  XXXXXXXXXXXXXXXXX",
"X    X    XXXXXXXXXXXXXXX",
"XE                XXXXXXX",
"XXXXXXXXXXXXXX    XXX   X",
"XXX  XXXXXXXE     XXX   X",
"XXX                     X",
"XXXXXXXXXXXX     XXXXXXXX",
"XX        XX   XXXXXXXXXX",
"XX   XE   XX           XX",
"XX   XX   XXXXXXXXXX   XX",
"XXX   X                XX",
"XXXXXXX   XXXXXXXXXX  XXX",
"XXX     XXXX   XXXX   XXX",
"XX  XX  XXXX   XXX   XXXX",
"XX  XX  XXXX   XX    XXX",
"XX   XXXXXXXE         XXX",
"XXX       TXXXXXXXXXXXXXX",
"XXXXXXXXXXXXXXXXXXXXXXXXX"
]

# ADD treasures list
treasures = []

enemies = [] 

# ADD maze to mazes list
levels.append(level_1)


# 미로 설정
def setup_maze(level):
    for y in range(len(level)):
        for x in range(len(level[y])):
            character = level[y][x]
            screen_x = -288 + (x * 24)
            screen_y = 288 - (y * 24)
           
            # "X" = 벽
            if character == "X":
                pen.goto(screen_x, screen_y)
                pen.stamp()
                walls.append((screen_x, screen_y))
            # "P" = 플레이어
            if character == "P":
                player.goto(screen_x, screen_y)
            # "T" = 보상  
            if character == "T":
                treasures.append(Treasure(screen_x, screen_y))
            # "E" = 적    
            if character == "E":
                enemies.append(Enemy(screen_x, screen_y))
                
            
# creat class instances                
pen = Pen()
player = Player()

# create wall list
walls = []

# set up the level
setup_maze(levels[1])

# keyboard 설정
turtle.listen()
turtle.onkey(player.go_left, "Left")
turtle.onkey(player.go_right, "Right")
turtle.onkey(player.go_up, "Up")
turtle.onkey(player.go_down, "Down")


# screen update
wn.tracer(0)

# start moving enemy
for enemy in enemies:
    turtle.ontimer(enemy.move, t=250)

# main game loop
while True:
    for treasure in treasures:
        if player.is_collision(treasure):
            player.gold += treasure.gold
            print("Player Gold: {}".format(player.gold))
            treasure.destroy()
            treasures.remove(treasure)
            
            
    for enemy in enemies:
        if player.is_collision(enemy):
            player.destroy()
            print("YOU DIED")
    
    # update screen
    wn.update()
```
